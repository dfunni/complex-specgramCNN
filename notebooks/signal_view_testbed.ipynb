```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

from scipy import signal
from scipy.fft import fftshift

%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
filename = "datasets/RML2016_10a_dict.pkl"
raw_data = pickle.load(open(filename, 'rb'), encoding='latin1')
```

```python
def plot_spectrogram(f, t, sxx):
#     f = np.linspace(0, 2*np.pi, f.size)
    plt.pcolormesh(t, 
                   fftshift(f), 
                   fftshift((sxx), axes=0))    
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time')
#     plt.show()
```

```python
f, t, sxx = signal.spectrogram(x[10,0], nperseg=29, noverlap=28, nfft=100, return_onesided=False, mode='complex')

fig = plt.figure()

plt.subplot(141)
plt.title('In-phase')
plot_spectrogram(f, t, np.real(sxx))

plt.subplot(142)
plt.title('Quadrature')
plot_spectrogram(f, t, np.imag(sxx))

plt.subplot(143)
plt.title('Magnitude')
plot_spectrogram(f, t, np.abs(sxx))

plt.subplot(144)
plt.title('Phase Angle')
plot_spectrogram(f, t, np.angle(sxx))

# plt.subplot(155)
# plt.title('Unwrapped Phase Angle')
# plot_spectrogram(f, t, np.unwrap(np.angle(sxx), axis=0))

fig.tight_layout()
plt.show()

```
