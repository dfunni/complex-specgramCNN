```python
def normalize_spectrogram(x_s):
    # normalizes spectrogram between 0 and 1
    # used in 2 channel iq_to_spec
    x_s[:,:,:,0] = (x_s[:,:,:,0]-np.min(x_s[:,:,:,0])) / (np.max(x_s[:,:,:,0])-np.min(x_s[:,:,:,0]))
    x_s[:,:,:,1] = (x_s[:,:,:,1]-np.min(x_s[:,:,:,1])) / (np.max(x_s[:,:,:,1])-np.min(x_s[:,:,:,1]))
    return x_s


def iq2specMP(data_dict, nperseg, noverlap, n_ex=None, nfft=None, unwrap=False):
    # creates spectrogram dict with three channels for psd, angle, and phase
    
    if n_ex == None:
        n_ex = data_dict['y'].size

    if nfft == None:
        n_f = nperseg
    else:
        n_f = nfft

    n_t = int((128-nperseg)/(nperseg-noverlap) + 1)
    
    x_s = np.zeros((n_f * n_ex, n_t, 2))
    y = []
    snr = []

    for i in range(n_ex):
        x = data_dict['x'][i]
        f, t, sxx = scipy.signal.spectrogram(x,
                                             window='hann',
                                             nperseg=nperseg,
                                             noverlap=noverlap,
                                             nfft=nfft,
                                             mode='complex',
                                             return_onesided=False)
        
        mag = np.abs(sxx)
        phase = np.angle(sxx)
        
        x_s[i*n_f : n_f*(i + 1), :, 0] = mag
        if unwrap:
            x_s[i*n_f : n_f*(i + 1), :, 1] = np.unwrap(phase)
        else:
            x_s[i*n_f : n_f*(i + 1), :, 1] = phase
        
        y.append(data_dict['y'][i])
        snr.append(data_dict['snr'][i])

    x_s = x_s.reshape(n_ex, n_f, n_t, 2)
    y = np.array(y)
    snr = np.array(snr)

    x_s = normalize_spectrogram(x_s)

    spec_dict = {'x_s': x_s, 'y': y, 'snr': snr, 't': t, 'f': f}
                
    return spec_dict


def iq2specIQ(data_dict, nperseg, noverlap, n_ex=None, nfft=None):
    # creates spectrogram dict with three channels for psd, angle, and phase
    
    if n_ex == None:
        n_ex = data_dict['y'].size

    if nfft == None:
        n_f = nperseg
    else:
        n_f = nfft

    n_t = int((128-nperseg)/(nperseg-noverlap) + 1)
    
    x_s = np.zeros((n_f * n_ex, n_t, 2))
    y = []
    snr = []

    for i in range(n_ex):
        x = data_dict['x'][i]
        f, t, sxx = scipy.signal.spectrogram(x,
                                             window='hann',
                                             nperseg=nperseg,
                                             noverlap=noverlap,
                                             nfft=nfft,
                                             mode='complex',
                                             return_onesided=False)
        
        x_s[i*n_f : n_f*(i + 1), :, 0] = np.real(sxx)
        x_s[i*n_f : n_f*(i + 1), :, 1] = np.imag(sxx)
        
        y.append(data_dict['y'][i])
        snr.append(data_dict['snr'][i])

    x_s = x_s.reshape(n_ex, n_f, n_t, 2)
    y = np.array(y)
    snr = np.array(snr)

    x_s = normalize_spectrogram(x_s)

    spec_dict = {'x_s': x_s, 'y': y, 'snr': snr, 't': t, 'f': f}
                
    return spec_dict


def plot_specMP(spec_dict, ex_num):
    # plots spectrogram for three channel dict version
    mag = spec_dict['x_s'][ex_num,:,:,0]
    phase = spec_dict['x_s'][ex_num,:,:,1]
    t = spec_dict['t']
    f = spec_dict['f']
    y = spec_dict['y'][ex_num]
    snr = spec_dict['snr'][ex_num]

    plt.subplot(121)
    mag = np.reshape(mag, (mag.shape[0], mag.shape[1]))
    plt.title('Magnitude\nWaveform: ' + y + '    SNR: ' + str(snr))
#     plt.pcolormesh(10*np.log10(psd))
    plt.pcolormesh(t, 
                   fftshift(f), 
                   fftshift((mag), 
                   axes=0), 
                   cmap='jet')
    
    plt.subplot(122)
    angle = np.reshape(phase, (phase.shape[0], phase.shape[1]))
    plt.title('Angle\nWaveform: ' + y + '    SNR: ' + str(snr))
#     plt.pcolormesh(10*np.log10(angle))
    plt.pcolormesh(t, 
                   fftshift(f), 
                   fftshift((angle), 
                   axes=0), 
                   cmap='jet')
    
    
    plt.show()
    
    
def plot_specIQ(spec_dict, ex_num):
    # plots spectrogram for three channel dict version
    I = spec_dict['x_s'][ex_num,:,:,0]
    Q = spec_dict['x_s'][ex_num,:,:,1]
    t = spec_dict['t']
    f = spec_dict['f']
    y = spec_dict['y'][ex_num]
    snr = spec_dict['snr'][ex_num]

    plt.subplot(121)
    mag = np.reshape(I, (I.shape[0], I.shape[1]))
    plt.title('In-Phase\nWaveform: ' + y + '    SNR: ' + str(snr))
#     plt.pcolormesh(10*np.log10(psd))
    plt.pcolormesh(t, 
                   fftshift(f), 
                   fftshift((mag), 
                   axes=0), 
                   cmap='jet')
    
    plt.subplot(122)
    angle = np.reshape(Q, (Q.shape[0], Q.shape[1]))
    plt.title('Quadrature\nWaveform: ' + y + '    SNR: ' + str(snr))
#     plt.pcolormesh(10*np.log10(angle))
    plt.pcolormesh(t, 
                   fftshift(f), 
                   fftshift((angle), 
                   axes=0), 
                   cmap='jet')
    
    
    plt.show()
```

```python
def interleave(x_c):
    I = np.real(x_c)
    Q = np.imag(x_c)
    x_iq = np.empty(I.size + Q.size, dtype=np.float32)
    x_iq[0::2] = I
    x_iq[1::2] = Q
    return x_iq

x_iq = interleave(x_c)
x_iq.shape
```

```python
def _normalize(arr):
    return (arr - arr.min()) / (arr.max() - arr.min())


def _sigma_thresh(arr, thresh):
    low_thresh, up_thresh = thresh
    sigma = np.std(arr)
    mx = np.mean(arr)
    arr = np.minimum(arr, mx + up_thresh*sigma)
    arr = np.maximum(mx - low_thresh*sigma, arr)
    return arr


def calc_spectrogram(x_c, fs, thresh, mode, nperseg, noverlap, nfft):
    f, t, sxx = signal.spectrogram(x_c, 
                                   fs, 
                                   nperseg=nperseg, 
                                   noverlap=noverlap, 
                                   nfft=nfft, 
                                   mode=mode, 
                                   return_onesided=False)

#     sxx = _sigma_thresh(sxx, thresh)
#     sxx = _normalize(sxx)

    return f, t, sxx


def plot_spectrogram(f, t, sxx, center_freq):
    offset = center_freq - f[-1]

    plt.pcolormesh(t, 
                   fftshift(f) + offset, 
                   fftshift((sxx), 
                   axes=0), 
                   cmap='jet')#,
                   # norm=LogNorm(.1, .5))

    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [s]')
    plt.show()
```
